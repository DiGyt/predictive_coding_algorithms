# 1. Correlation

A. Get the prior correlations

```python
tf.random.set_seed(15042024)

# decide whether to project to input level or compare prior inc
to_input_level = False

# get dataset
for x, y in val_dataset.batch(99999).take(1):
  x, y = EXP_PARADIGMS[exp_paradigm](x, y, 0., y_return="y")

# Initialize an empty DataFrame with the desired columns
prior_df = pd.DataFrame(columns=["Training Objective", "Correlation", "Lower Error", "Upper Error"])

# get prior correlations
for key, tmp_model in models_dict.items():

    # get priors
    preds = tmp_model(x)
    priors = preds[:, :-1] @ tmp_model.layers[-1].cell.recurrent_kernel

    if to_input_level:
        # use reconstructed pred
        priors = model_reconstruct_prediction(tmp_model, priors)[0]
        correlation = np.corrcoef(np.reshape(priors, [-1]), -np.reshape(x[:, 1:], [-1]))[0, 1]
    else:
        # use prior-input corr
        drives = get_input_drive(tmp_model.layers[-1], preds[:, 1:])
        correlation = np.corrcoef(np.reshape(priors, [-1]), -np.reshape(drives, [-1]))[0, 1]

    # Calculate confidence intervals
    llci, ulci = fisher_z_confidence_interval(correlation, len(x))
    lower_error, upper_error = correlation - llci, ulci - correlation

    # Append the new row to the DataFrame
    prior_df.loc[len(prior_df)] = {
    "Training Objective": key,
    "Correlation": correlation,
    "Lower Error": lower_error,
    "Upper Error": upper_error}

plot_input_series(x, "Input Example", sample_idx=0, show_indices=[0, 2, 4, 6, 8, 10])
```

```python
# Prepare a DataFrame to store p-values
p_values = pd.DataFrame(columns=["Comparison", "T statistic", "significance level", "corrected significance level"])

# Do bonferroni correction
n_comparisons = len(pairs)

# compare correlations betwee each pair
for pair in pairs:
    r1 = prior_df["Correlation"][prior_df["Training Objective"] == pair[0]].iloc[0]
    r2 = prior_df["Correlation"][prior_df["Training Objective"] == pair[1]].iloc[0]
    t_diff, p_value = compare_correlations(r1, len(x), r2, len(x))
    p_values.loc[len(p_values)] = {
    "Comparison": pair,
    "T statistic": t_diff,
    "significance level": p_value,
    "corrected significance level": np.minimum(p_value * n_comparisons, 1)
}
```

C. Plot

```python
# Create a list to store custom annotations and significant pairs
custom_annotations = []
significant_pairs = []

# assign significance values for each comparison
for pair in pairs:
    p_value = p_values[p_values["Comparison"] == pair]["corrected significance level"].iloc[0]
    if p_value < 0.05:  # Adjust significance level as needed
        significant_pairs.append(pair)
        if p_value < 0.0001:
            custom_annotations.append("****")
        elif p_value < 0.001:
            custom_annotations.append("***")
        elif p_value < 0.01:
            custom_annotations.append("**")
        else:
            custom_annotations.append("*")

# plot the data
plot_input_series(x, "Input Example", sample_idx=0, show_indices=[0, 2, 4, 6, 8, 10])
plt.figure(figsize=(10, 5), dpi=100)
if True:
    ax = sns.barplot(data=prior_df, x='Training Objective', y='Correlation', order=order, palette=my_colors)
    plt.errorbar(x=np.arange(len(prior_df)), y=prior_df['Correlation'],
             yerr=[prior_df['Lower Error'], prior_df['Upper Error']],
             fmt='none', ecolor='black', capsize=5)
    plt.xlabel("Training Objective")
    if to_input_level:
      plt.ylabel("Correlation of Prior(t) with True Input t(+1)")
    else:
      plt.ylabel("Correlation of Prior(t) with bottom-up signal t(+1)")

    # Create Annotator and set custom annotations
    annotator = Annotator(ax, significant_pairs, data=prior_df, x='Training Objective', y='Correlation', order=order)
    annotator.set_custom_annotations(custom_annotations)
    annotator.configure(text_format='star', loc='outside', verbose=2, text_offset=1, line_offset=-5, line_offset_to_group=-5, line_width=0.5, line_height=0.01, fontsize=8)  # Adjust parameters as needed


    # Apply annotations
    annotator.annotate()
    plt.ylim([None, 1])


    # Save the current figure as SVG
    output_name = "prior_input" if to_input_level else "prior_encoded"
    output_path = os.path.join(figures_dir, output_name)
    plt.savefig(output_path + ".svg", format='svg')
    plt.savefig(output_path + ".png", format='png')

    plt.show()
```

```python
# define the experiment sequence
tf.random.set_seed(15042024)

# true label
conditions = {"Low Noise":(lambda x,y:add_noise(x, y, 0.0)),
              "High Noise":(lambda x,y:add_noise(x, y, 0.5))}

# get dataset
condition_data = {}
for x, y in val_dataset.batch(99999).take(1):
    x, y = EXP_PARADIGMS[exp_paradigm](x, y, 0., y_return="y")
    for cond, cond_preproc in conditions.items():
        condition_data[cond], _ = cond_preproc(x, y)
```

```python
# set seed
tf.random.set_seed(15042024)

# define models to look at
algorithms = ["Contrastive", "Predictive", "Supervised"]
cond_dict = {"Contrastive":"Forward_Forward", "Predictive":"Predictive_Prior", "Supervised":"Supervised"}
regularizations = ["No Regularization", "Regularization"]

# go through all conditions and extract activations
activations = {}
for model_class in algorithms:
  activations[model_class] = {}
  for regularization in regularizations:
      print("CHECKING MODEL NOW: ", model_class, regularization)
      activations[model_class][regularization] = {}

      # load a specific model
      reg = True if regularization == "Regularization" else False
      fname = f"exp_Statistical_Learning-algo_{cond_dict[model_class]}-loc_False-rec_True-gain_False-sparse_{reg}"
      model = tf.keras.saving.load_model(op.join(save_dir, fname, "model"), custom_objects=custom_objects)

      # extract activations
      for cond, x_cond in condition_data.items():
        act = []
        out = x_cond
        for layer in model.layers:
          if isinstance(layer, (tf.keras.layers.SimpleRNN, FrozenSimpleRNN)):
            outs = layer(out)
            act.append(outs[:, :1] - np.mean(outs[:, :1]))
          out = layer(out)
        acts = tf.concat([l for l in act], axis=-1)[:, 0, :]

        # store activations
        activations[model_class][regularization][cond] = acts.numpy()
```

```python
# define number of comparisons
n_comparisons = len(algorithms) * 2
stats_df = pd.DataFrame(columns=["Algorithm", "Variance ratio W", "Variance ratio Z", "Variance ratio P", "Variance ratio P (corrected)", "Regularization W", "Regularization Z", "Regularization P", "Regularization P (corrected)"])

# calculate variance ratio for all algorithms
for algorithm in algorithms:
  W_vr = stats.wilcoxon(activations[algorithm]["No Regularization"]["High Noise"].var(-1) / activations[algorithm]["No Regularization"]["Low Noise"].var(-1) -\
                        activations[algorithm]["Regularization"]["High Noise"].var(-1) / activations[algorithm]["Regularization"]["Low Noise"].var(-1), method="approx")

  W_reg = stats.wilcoxon(activations[algorithm]["Regularization"]["High Noise"].var(-1) - activations[algorithm]["Regularization"]["Low Noise"].var(-1), method="approx")

  # fill dataframe with all relevant values
  stats_df.loc[len(stats_df)] = {
    "Algorithm": algorithm,
    "Variance ratio W": W_vr.statistic,
    "Variance ratio Z": W_vr.zstatistic,
    "Variance ratio P": W_vr.pvalue,
    "Variance ratio P (corrected)": np.minimum(W_vr.pvalue * n_comparisons, 1),
    "Regularization W": W_reg.statistic,
    "Regularization Z": W_reg.zstatistic,
    "Regularization P": W_reg.pvalue,
    "Regularization P (corrected)": np.minimum(W_reg.pvalue * n_comparisons, 1)
}

```
